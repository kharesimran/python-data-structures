```python
# append and extend
b = a.append(6)
print(a)
print(b) # Nothing is returned. List is modified.
c = a.extend([7, 8]) # iterable
print(a)
print(c) # Nothing is returned. List is modified.
```

```python
# indexing and splicing
a = [1, 2, 3, 4, 5, 6, 7, 8, 6, 7, 8]
print(a)
print(a[4])
print(a[4:7])
print(a[-1])
print(a[-2])
print(a[-11])
print(a[len(a) - 1])
print(a[:])
print(a[:-2])
print(a[4:])
print(a[4:-2])

print('index function')
i = [1, 2, 3, 4, 3, 6, 7, 8, 6, 7, 8]
print(i)
b = i.index(3)
print(b)
```

```python
# remove, pop and del
a = [1, 2, 3, 4, 5, 6, 7]
elem  = 2
b = a.remove(elem)
print(a)
print(b) # Returns None
c = a.pop()
print(a)
print(c)
idx = 1
d = a.pop(idx)
print(a)
print(d)
del a[1]
print(a)
```

```python
# assignment (only pointer copied), copy (shallow and deep), copy using splicing
print('Assignment')
a = [1, 2, 3, 4, 5, 6, 7]
b = a
a[0] = 0
print(a, id(a))
print(b, id(b)) # Also changes

print('Copy')
import copy
a = [1, 2, 3, 4, 5, 6, 7]
b = a.copy() # Shallow copy - same as deep copy as list is not nested
c = copy.copy(a)
a[0] = 0
print(a, id(a))
print(b, id(b)) # Does not change
print(c, id(c))

# Nested list
print('Nested')
a = [[1, 2, 3], [4, 5, 6], 7]
b = a.copy() # Shallow copy - same as deep copy as list is not nested
c = copy.copy(a)
a[0][0] = 0
print(a, id(a[0]))
print(b, id(b[0])) 
print(c, id(c[0]))

print('Deep copy')
a = [[[1, 5], 2, 3], [4, 5, 6], 7]
b = a.copy() # Shallow copy - same as deep copy as list is not nested
c = copy.deepcopy(a)
a[0][0][0] = 0
print(a, id(a[0]), id(a[0][0]))
print(b, id(b[0]), id(b[0][0])) 
print(c, id(c[0]), id(c[0][0]))
```

```python
print('a.sort')
a = [1, 9, 5, 4, 2, 3, 6]
print(a)
b = a.sort() # Important! It is a.sort() not 'ed'
print(b)
print(a)

print('sorted(a)')
a = [1, 9, 5, 4, 2, 3, 6]
print(a)
b = sorted(a)
print(b)
print(a) # Unchanged

print('Reverse sort')
print(a)
b = a.sort(reverse=True)
print(b)
print(a)
```

```python
print('Sorting with key')
a = ['hi', 'longwordhere', 'am', 'four', 'hellothere']
print(a)
a.sort()
print(a) # alphabetically
a.sort(key= lambda x: len(x))
print(a)
a.sort(key= lambda x: len(x), reverse=True)
print(a)
```
